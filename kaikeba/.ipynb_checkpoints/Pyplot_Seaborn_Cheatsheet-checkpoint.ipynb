## 子图 - subplot

1. 将一个画布进行切分（Figure）默认每次画的时候都是一个figure

2. 将切分后的图分配到固定的位置

3. 将图可以设置成固定的大小

```python
# 栅格系统，从左往右进行计数

plt.figure(1) # figure的name是1
plt.subplot(131) # 一行三列，位于第一个位置
plt.bar(names,value,color='g')
plt.subplot(235) # 两行三列，位于第五个位置
plt.scatter(names,value,color='y')
plt.subplot(233) # 两行三列，位于第三个位置
plt.plot(names,value,color='r')
plt.title('离散数据的柱状图、散点图、折线图')

# 如果有图重复覆盖，则原先的图被后面的图覆盖
```

## Seaborn的绘图练习

道/ 法/ 术/ 器

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

## 1. 带状图-离散数据和连续数据之间的关系

```python
sns.stripplot(data=tips, x='day',y='total_bill')
```

```python
sns.stripplot(data=tips, x='day',y='total_bill',jitter=False) 
# jitter是随机抖动，设成False就是完全的带状，点允许重合
```

## 2. 蜂群图-离散数据和连续数据的关系-密度排列

```python
# 带状图的补充，如果数据相差较近，会横向排列，点之间不允许重合
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
tips.head()
```

分析每天中，午餐和晚餐的账单分布

## 3. Hue-分组参数

加了Hue相当于一个三元组。原来的X，y不变，加了Hue相当于又加了一个变量

在每天的付账人群中午餐、晚餐分布

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='time')
```

在每天的付账人群中的性别分布

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex')
```

披萨的大小

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='size')
```

```python
# 揭示了，Pizza的不同size的基础价格
sns.swarmplot(x='size',y='total_bill',data=tips)
```

```python
tips['size'].corr(tips['total_bill'])
```

## 4. 箱线图

表示统计的规律

```python
sns.boxplot('day','total_bill',data=tips)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot('day','total_bill',data=tips,hue='time')
```

## 5. 提琴图

表示total_bill的概率分布

```python
sns.violinplot('day','total_bill',data=tips,hue='time')
```

可以看到概率分布整体下偏

```python
sns.violinplot('day','total_bill',data=tips,hue='time',split=True)
```

多图合成展示

```python
sns.violinplot('day','total_bill',data=tips)
sns.swarmplot('day','total_bill',data=tips,color='w')
```

## 带回归的散点图

```python
# 加一个线性回归，将皮尔森系数算出来，带影子的部分是置信区间，也就是说将他的方差（斜率估计）计算出来了
sns.lmplot('size','total_bill',data=tips)
```

## 联合分布图（既有散点的特性，又有线性回归，同时还有概率估计）

```python
sns.jointplot('total_bill','tip',data=tips,kind='reg')
```
