## 算法的Baseline

算法的pipeline：

数据采集 -> 数据清洗 -> 数据仓库（存储） -> 数据挖掘 -> 数据标注 -> 数据集市（训练集/评测集）-> 训练 -> 评测 -> 
    模型工程（压缩、剪枝、Int8定点化等） -> 模型的集成和应用


算法的Baseline：

训练集的制作 -> 训练 -> 评测

在形成一个Baseline之后，模型迭代的过程 进步的标志 是 评测指标的提升

评测指标：

【AUC, ROC, AP(Average Precision平均查准率，是PR曲线的面积)】

一般要先分Regression 还是 Classification

Regression

   MSE(Mean Squared Error)--L2 / MAE(Mean Absolute Error)--L1

Classification

   Cross-Entropy(交叉熵)
   
   信息量与概率成反比，熵与混乱程度也有关系，熵越大越混乱

   信息量 -$log_2(P)$ -> 信息熵 -$\Sigma(p * log_2(p))$ -> 交叉熵 -$\Sigma(p * log_2(q))$ 
   
   交叉熵中的q就是自己的预测，p是真实的情况，所有交叉熵就是以个人肤浅的眼光去看世界中真实存在的事物
   
   Center-Loss

### 缺失值的类型

Numerical  |  Categorical

#### Numerical

1. 均值填充

2. 中位数填充

3. 高频填充 -- 用的会“稍微”少一点 -- 连续型数据本身是概率密度，很难数字相等，但是很容易接近

4. 分布拟合填充 -- 回归一个分布，或者是（$\mu$ - 均值， $\sigma$ - 标准差）- 在（+- $\sigma$）范围内进行随机填充

5. 内在关系填充（根据列本身存在的意义和数据集内其他列的相关性构建数学映射模型进行填充）

#### Categorical

1. 高频填充

"男" | "女" --> 0 | 1

"小孩" | "青年" | "中年" | "老年" --> 0 | 1 | 2 | 3

* No Free Lunch Theory -- 不存在超级机器学习模型

如果不付出努力，没有好的机器学习模型

2. 内在关系填充

## 交叉验证

1. 数据有限 -> 发挥数据本来的效率
2. 数据的训练集和评测集的矛盾。 

*如果用更多数据训练，会有更少的数据来评测 -> 失去了对于真实泛化能力的考量
 
*如果用更少的数据来训练，很大可能造成欠拟合，在评测集熵表现就一定不好

解决办法：

1. 将数据集拆成k份 -> 首先进行shuffle

2. 规定k-1份进行训练，剩下的1份进行评测，总共训练k次，轮流每个子数据集作为评测集

*数据被分成K份， 1.。。。。K

第一次训练使用第1份数据集作为评测集，剩余的k-1份作为训练集

第i次训练使用第i份数据集作为评测集，剩余的k-1份作为训练集

做k次训练

3. k次训练之后，评测的分值=k次评测结果的平均

4. 不同的模型会提高不同的子模型的合并方法，会将所有的k个子模型进行合并

|r| > 0.95 显著关系 *显著的关系需要考虑该特征是否与目标有很强的关联关系（如 报价 和 成交价，这样可以考虑去掉，否则影响模型泛化能力）

|r| < 0.3 极弱相关

一张图片上画了一只猫：

p_dog = 0

q_cat = 0.2

q_dog = 0.8

```python
garage_cols=['GarageType','GarageQual','GarageCond','GarageYrBlt','GarageFinish','GarageCars','GarageArea']
```

```python
train[garage_cols][train['GarageType'].isnull()]
# 都是空，都是NaN，所以可以填成空
```

教给了我们一个填充技巧，就是用相关特征来进行分析

```python
for col in garage_cols:
    if train[col].dtype == np.object:
        train[col]=train[col].fillna('None')
    else:
        train[col] = train[col].fillna(0)
```
